```python
def implies(p, q):
    return not p or q

def contrapositive(p, q):
    return implies(not q, not p)

def converse(p, q):
    return implies(q, p)

def inverse(p, q):
    return implies(not p, not q)

def biconditional(p, q):
    return p == q

def xor(p,q):
    return (p or q) and not ( (p and q))

# Logical equivalences
def distributive(p, q, r):
    return implies(p or q, r) == (implies(p, r) and implies(q, r))

def exportation(p, q, r):
    return implies(p and q, r) == implies(p, implies(q, r))

def reduction(p, q):
    return not implies(p, q) == (p and not q)

def equivalence(p,q):
    return biconditional(p,q) == (not xor(p,q))

def truthTable():
    # Columns
    # TASK: Modify each column as required for your compound statement (p, q, r, p ∨ q, (p ∧ q) → r, p → q, (p → r) ∧ (q → r))
    
    col1 = 'p'
    col2 = 'q'
    col3 = 'r'
    col4 = 'p ∨ q'
    col5 = '(p ∧ q) → r'
    col6 = 'p → q'
    col7 = '(p → r) ∧ (q → r)'
    print(f'{col1} \t{col2} \t{col3} \t{col4} \t{col5} \t{col6} \t{col7}')
    print('-'*100)
    
    # Iteration of proposition values
    for p in [True, False]:
        for q in [True, False]:
            for r in [True, False]:
                # Columns
                # TASK: Modify each column as required for your compound statement
                col1 = p
                col2 = q
                col3 = r
                col4 = p or q
                col5 = implies((p and q), r)
                col6 = implies(p,q)
                col7 = implies(p,r) and implies(q,r)
                print(f'{col1} \t{col2} \t{col3} \t{col4} \t{col5} \t{col6} \t{col7}')

truthTable()

#Test Cases
p, q, r = True, False, True
print("p is True, q is False, and r is True")
print(f"p → q {implies(p, q)}")  
print(f"The Contrapositive(p → q): {contrapositive(p, q)}")
print(f"The Converse(p → q): {converse(p, q)}")
print(f"The Inverse (p → q): {inverse(p, q)}")
print(f"The Biconditional (p bi q): {biconditional(p, q)}")
print(f"The XOR (p xor q): {xor(p, q)}")

print(f"Distributiv: {distributive(p, q, r)}")
print(f"Exportation: {exportation(p, q, r)}")
print(f"Reduction: {reduction(p, q)}")
print(f"Equivalence: {equivalence(p, q)}")

#English Statements
# p = I am late to work, q = I ran to work
# Original, if I am late to work, I will run to work
# Inverse, if I am not late to work, I will not run to work
# Converse, I will run to work, If I am late to work
# Contrapositive, I will not run to work if I am not late to work
```
